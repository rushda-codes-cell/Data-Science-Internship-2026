```python
import time
from prometheus_client import Counter, Histogram, start_http_server

# Start server only if port 8000 isn't already listening
try:
    start_http_server(8000)
    print("Metrics server started at http://localhost:8000")
except OSError:
    print("Metrics server is already running on http://localhost:8000")
```

```python
from prometheus_client import REGISTRY, Counter, Histogram

# Reuse existing metrics if already declared in kernel session
if "retailpulse_app_visits" not in REGISTRY._names_to_collectors:
    APP_VISITS = Counter(
        "retailpulse_app_visits", "Total notebook executions"
    )
    MODEL_TIME = Histogram(
        "model_execution_time_seconds", "Model execution time"
    )
    print("Custom Metrics Created")
else:
    print("Metrics already initialized in Registry")
```

```python
import time

# Increment visit count
APP_VISITS.inc()

# Record simulated execution time
with MODEL_TIME.time():
    time.sleep(1.5)

print("Logged metric update!")
```
