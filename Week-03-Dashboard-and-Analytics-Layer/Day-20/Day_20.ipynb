```python
import streamlit as st
import pandas as pd
from reportlab.lib.styles import getSampleStyleSheet
from reportlab.platypus import SimpleDocTemplate, Paragraph
```

```python
forecast = pd.read_csv("hybrid_forecast_results.csv")
inventory = pd.read_csv("inventory_optimization_results.csv")
churn = pd.read_csv("customer_churn_predictions.csv")
```

```python
st.title("📥 Export Reports")
```

```python
st.subheader("Export Forecast Data")

csv = forecast.to_csv(index=False)

st.download_button(
    label="Download Forecast CSV",
    data=csv,
    file_name="forecast_report.csv",
    mime="text/csv"
)
```

```python
st.subheader("Export Inventory Data")

csv = inventory.to_csv(index=False)

st.download_button(
    label="Download Inventory CSV",
    data=csv,
    file_name="inventory_report.csv",
    mime="text/csv"
)
```

```python
st.subheader("Export Customer Churn Data")

csv = churn.to_csv(index=False)

st.download_button(
    label="Download Churn CSV",
    data=csv,
    file_name="customer_churn_report.csv",
    mime="text/csv"
)
```

```python
def create_pdf():

    pdf_file = "RetailPulse_Report.pdf"

    doc = SimpleDocTemplate(pdf_file)

    styles = getSampleStyleSheet()

    story = []

    story.append(Paragraph("RetailPulse Report", styles["Title"]))

    story.append(Paragraph("<br/>", styles["BodyText"]))

    story.append(Paragraph(
        f"Total Forecast : {forecast['Hybrid_Prediction'].sum():.2f}",
        styles["BodyText"]
    ))

    story.append(Paragraph(
        f"Average Inventory : {inventory['Recommended_Inventory'].mean():.2f}",
        styles["BodyText"]
    ))

    story.append(Paragraph(
        f"Average Churn Probability : {churn['ChurnProbability'].mean()*100:.2f}%",
        styles["BodyText"]
    ))

    story.append(Paragraph(
        "Report Generated Successfully.",
        styles["BodyText"]
    ))

    doc.build(story)

    return pdf_file
```

```python
if st.button("Generate PDF Report"):

    pdf = create_pdf()

    with open(pdf, "rb") as file:

        st.download_button(
            label="Download PDF Report",
            data=file,
            file_name="RetailPulse_Report.pdf",
            mime="application/pdf"
        )
```
