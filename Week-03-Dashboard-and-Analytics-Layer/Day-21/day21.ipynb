```python
%%writefile app.py

import streamlit as st
import pandas as pd
from datetime import datetime

from reportlab.lib.styles import getSampleStyleSheet
from reportlab.platypus import SimpleDocTemplate, Paragraph


# =====================================================
# PAGE CONFIGURATION
# =====================================================

st.set_page_config(
    page_title="RetailPulse Dashboard",
    page_icon="📊",
    layout="wide"
)


# =====================================================
# LOAD DATA
# =====================================================

@st.cache_data
def load_data():

    forecast = pd.read_csv(
        "hybrid_forecast_results.csv"
    )

    prophet = pd.read_csv(
        "prophet_forecast.csv"
    )

    inventory = pd.read_csv(
        "inventory_optimization_results.csv"
    )

    churn = pd.read_csv(
        "customer_churn_predictions.csv"
    )

    rfm = pd.read_csv(
        "rfm_features.csv"
    )

    rolling = pd.read_csv(
        "rolling_statistics.csv"
    )

    return (
        forecast,
        prophet,
        inventory,
        churn,
        rfm,
        rolling
    )


forecast, prophet, inventory, churn, rfm, rolling = load_data()


# =====================================================
# SIDEBAR
# =====================================================

st.sidebar.title(
    "🔍 Dashboard Filters"
)


# Churn Threshold

churn_threshold = st.sidebar.slider(
    "Churn Risk Threshold",
    min_value=0.0,
    max_value=1.0,
    value=0.80,
    step=0.05
)


# Inventory Status

if "Inventory_Status" in inventory.columns:

    inventory_status_options = [
        "All"
    ] + sorted(
        inventory[
            "Inventory_Status"
        ]
        .dropna()
        .unique()
        .tolist()
    )

    selected_inventory_status = st.sidebar.selectbox(
        "Inventory Status",
        inventory_status_options
    )

else:

    selected_inventory_status = "All"


# Demand Scenario

scenario = st.sidebar.selectbox(
    "Demand Scenario",
    [
        "Normal Demand",
        "High Demand (+20%)",
        "Low Demand (-20%)"
    ]
)


# =====================================================
# FILTER DATA
# =====================================================


filtered_churn = churn[
    churn[
        "ChurnProbability"
    ] >= churn_threshold
]


if selected_inventory_status != "All":

    filtered_inventory = inventory[
        inventory[
            "Inventory_Status"
        ] == selected_inventory_status
    ]

else:

    filtered_inventory = inventory.copy()


# =====================================================
# HEADER
# =====================================================

st.title(
    "📊 RetailPulse - Real-Time Metrics & Alerts"
)


st.write(
    "Interactive Retail Analytics Dashboard"
)


st.success(
    "Last Updated: "
    + datetime.now().strftime(
        "%d-%m-%Y %H:%M:%S"
    )
)


# Refresh Button

if st.button(
    "🔄 Refresh Dashboard"
):

    st.cache_data.clear()

    st.rerun()


# =====================================================
# EXECUTIVE SUMMARY
# =====================================================

st.header(
    "📌 Executive Summary"
)


col1, col2, col3, col4 = st.columns(4)


col1.metric(
    "Total Forecast",
    f"{forecast['Hybrid_Prediction'].sum():,.0f}"
)


col2.metric(
    "Average Inventory",
    f"{inventory['Recommended_Inventory'].mean():,.0f}"
)


col3.metric(
    "Total Customers",
    f"{len(churn):,}"
)


col4.metric(
    "High Risk Customers",
    f"{len(filtered_churn):,}"
)


# =====================================================
# FORECAST METRICS
# =====================================================

st.header(
    "📈 Forecast Metrics"
)


col1, col2, col3, col4 = st.columns(4)


col1.metric(
    "Total Forecast",
    int(
        forecast[
            "Hybrid_Prediction"
        ].sum()
    )
)


col2.metric(
    "Average Forecast",
    round(
        forecast[
            "Hybrid_Prediction"
        ].mean(),
        2
    )
)


col3.metric(
    "Maximum Forecast",
    int(
        forecast[
            "Hybrid_Prediction"
        ].max()
    )
)


col4.metric(
    "Minimum Forecast",
    int(
        forecast[
            "Hybrid_Prediction"
        ].min()
    )
)


# =====================================================
# FORECAST ALERT
# =====================================================

if forecast[
    "Hybrid_Prediction"
].max() > 500:

    st.warning(
        "⚠ High Demand Detected"
    )

else:

    st.success(
        "✅ Demand is Normal"
    )


# =====================================================
# HYBRID FORECAST CHART
# =====================================================

st.subheader(
    "📊 Hybrid Forecast"
)


if "Date" in forecast.columns:

    forecast_chart = forecast.copy()

    forecast_chart[
        "Date"
    ] = pd.to_datetime(
        forecast_chart[
            "Date"
        ],
        errors="coerce"
    )

    forecast_chart = forecast_chart.dropna(
        subset=[
            "Date"
        ]
    )

    st.line_chart(
        forecast_chart.set_index(
            "Date"
        )[
            "Hybrid_Prediction"
        ]
    )

else:

    st.line_chart(
        forecast[
            "Hybrid_Prediction"
        ]
    )


# =====================================================
# PROPHET FORECAST
# =====================================================

st.subheader(
    "📈 Prophet Forecast"
)


if "yhat" in prophet.columns:

    st.line_chart(
        prophet[
            "yhat"
        ]
    )


# =====================================================
# WHAT-IF ANALYSIS
# =====================================================

st.header(
    "🔮 What-If Demand Analysis"
)


what_if = forecast.copy()


what_if[
    "Normal Demand"
] = what_if[
    "Hybrid_Prediction"
]


what_if[
    "High Demand (+20%)"
] = what_if[
    "Hybrid_Prediction"
] * 1.20


what_if[
    "Low Demand (-20%)"
] = what_if[
    "Hybrid_Prediction"
] * 0.80


st.subheader(
    "Selected Scenario: "
    + scenario
)


st.line_chart(
    what_if[
        scenario
    ]
)


# =====================================================
# INVENTORY METRICS
# =====================================================

st.header(
    "📦 Inventory Metrics"
)


col1, col2, col3 = st.columns(3)


col1.metric(
    "Average Inventory",
    round(
        inventory[
            "Recommended_Inventory"
        ].mean(),
        2
    )
)


col2.metric(
    "Maximum Inventory",
    int(
        inventory[
            "Recommended_Inventory"
        ].max()
    )
)


col3.metric(
    "Minimum Inventory",
    int(
        inventory[
            "Recommended_Inventory"
        ].min()
    )
)


# =====================================================
# INVENTORY ALERTS
# =====================================================

st.subheader(
    "⚠ Inventory Alerts"
)


low_stock = pd.DataFrame()


if "Inventory_Status" in inventory.columns:

    low_stock = filtered_inventory[
        filtered_inventory[
            "Inventory_Status"
        ] == "Low Stock"
    ]

    if len(low_stock) > 0:

        st.error(
            f"⚠ {len(low_stock)} products are Low Stock."
        )

    else:

        st.success(
            "✅ Inventory is Healthy."
        )


# =====================================================
# INVENTORY TABLE
# =====================================================

st.subheader(
    "📋 Inventory Recommendation"
)


st.dataframe(
    filtered_inventory,
    use_container_width=True
)


# =====================================================
# CUSTOMER CHURN
# =====================================================

st.header(
    "👥 Customer Churn Analysis"
)


col1, col2, col3 = st.columns(3)


col1.metric(
    "Total Customers",
    len(churn)
)


col2.metric(
    "Average Churn Probability",
    f"{churn['ChurnProbability'].mean() * 100:.2f}%"
)


col3.metric(
    "High Risk Customers",
    len(filtered_churn)
)


# =====================================================
# CUSTOMER ALERT
# =====================================================

st.subheader(
    "🚨 Customer Alerts"
)


if len(filtered_churn) > 0:

    st.error(
        f"⚠ {len(filtered_churn)} High Risk Customers Found"
    )

else:

    st.success(
        "✅ No High Risk Customers"
    )


# =====================================================
# CHURN CHART
# =====================================================

st.subheader(
    "📊 Churn Probability"
)


if len(filtered_churn) > 0:

    if "Customer ID" in filtered_churn.columns:

        churn_chart = filtered_churn[
            [
                "Customer ID",
                "ChurnProbability"
            ]
        ].head(20)

        churn_chart = churn_chart.set_index(
            "Customer ID"
        )

        st.bar_chart(
            churn_chart
        )

    else:

        st.bar_chart(
            filtered_churn[
                "ChurnProbability"
            ].head(20)
        )


# =====================================================
# CHURN TABLE
# =====================================================

st.subheader(
    "📋 Customer Churn Predictions"
)


st.dataframe(
    filtered_churn,
    use_container_width=True
)


# =====================================================
# CUSTOMER SEGMENTATION
# =====================================================

st.header(
    "🧩 Customer Segmentation"
)


col1, col2, col3 = st.columns(3)


col1.metric(
    "Customers",
    len(rfm)
)


if "Frequency" in rfm.columns:

    col2.metric(
        "Average Frequency",
        round(
            rfm[
                "Frequency"
            ].mean(),
            2
        )
    )


if "Monetary" in rfm.columns:

    col3.metric(
        "Average Monetary",
        round(
            rfm[
                "Monetary"
            ].mean(),
            2
        )
    )


st.subheader(
    "📋 RFM Customer Data"
)


st.dataframe(
    rfm,
    use_container_width=True
)


# =====================================================
# ROLLING STATISTICS
# =====================================================

st.header(
    "📉 Rolling Statistics"
)


rolling_columns = []


if "RollingMean7" in rolling.columns:

    rolling_columns.append(
        "RollingMean7"
    )


if "RollingStd7" in rolling.columns:

    rolling_columns.append(
        "RollingStd7"
    )


if len(rolling_columns) > 0:

    st.line_chart(
        rolling[
            rolling_columns
        ]
    )


# =====================================================
# VOLATILITY ALERT
# =====================================================

if (
    "RollingStd7" in rolling.columns
    and rolling[
        "RollingStd7"
    ].max() > 100
):

    st.warning(
        "⚠ Sales Volatility Increasing"
    )

else:

    st.success(
        "✅ Sales Stable"
    )


# =====================================================
# EXPORT REPORTS
# =====================================================

st.header(
    "📥 Export Reports"
)


# Forecast CSV

st.subheader(
    "📈 Export Forecast Data"
)


forecast_csv = forecast.to_csv(
    index=False
)


st.download_button(
    label="⬇ Download Forecast CSV",
    data=forecast_csv,
    file_name="forecast_report.csv",
    mime="text/csv"
)


# Inventory CSV

st.subheader(
    "📦 Export Inventory Data"
)


inventory_csv = inventory.to_csv(
    index=False
)


st.download_button(
    label="⬇ Download Inventory CSV",
    data=inventory_csv,
    file_name="inventory_report.csv",
    mime="text/csv"
)


# Churn CSV

st.subheader(
    "👥 Export Customer Churn Data"
)


churn_csv = churn.to_csv(
    index=False
)


st.download_button(
    label="⬇ Download Churn CSV",
    data=churn_csv,
    file_name="customer_churn_report.csv",
    mime="text/csv"
)


# =====================================================
# PDF REPORT
# =====================================================

st.subheader(
    "📄 Generate PDF Report"
)


def create_pdf():

    pdf_file = (
        "RetailPulse_Report.pdf"
    )


    doc = SimpleDocTemplate(
        pdf_file
    )


    styles = (
        getSampleStyleSheet()
    )


    story = []


    story.append(
        Paragraph(
            "RetailPulse Report",
            styles[
                "Title"
            ]
        )
    )


    story.append(
        Paragraph(
            "<br/>",
            styles[
                "BodyText"
            ]
        )
    )


    story.append(
        Paragraph(
            f"Total Forecast: "
            f"{forecast['Hybrid_Prediction'].sum():.2f}",
            styles[
                "BodyText"
            ]
        )
    )


    story.append(
        Paragraph(
            f"Average Inventory: "
            f"{inventory['Recommended_Inventory'].mean():.2f}",
            styles[
                "BodyText"
            ]
        )
    )


    story.append(
        Paragraph(
            f"Average Churn Probability: "
            f"{churn['ChurnProbability'].mean() * 100:.2f}%",
            styles[
                "BodyText"
            ]
        )
    )


    story.append(
        Paragraph(
            f"High Risk Customers: "
            f"{len(filtered_churn)}",
            styles[
                "BodyText"
            ]
        )
    )


    story.append(
        Paragraph(
            f"Low Stock Products: "
            f"{len(low_stock)}",
            styles[
                "BodyText"
            ]
        )
    )


    story.append(
        Paragraph(
            "Report Generated Successfully.",
            styles[
                "BodyText"
            ]
        )
    )


    doc.build(
        story
    )


    return pdf_file


if st.button(
    "📄 Generate PDF Report"
):

    pdf = create_pdf()


    with open(
        pdf,
        "rb"
    ) as file:

        st.download_button(
            label="⬇ Download PDF Report",
            data=file,
            file_name="RetailPulse_Report.pdf",
            mime="application/pdf"
        )
```
