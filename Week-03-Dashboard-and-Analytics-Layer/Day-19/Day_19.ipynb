```python
import pandas as pd
import streamlit as st
from datetime import datetime
```

```python
forecast = pd.read_csv("hybrid_forecast_results.csv")

prophet = pd.read_csv("prophet_forecast.csv")

inventory = pd.read_csv("inventory_optimization_results.csv")

churn = pd.read_csv("customer_churn_predictions.csv")

rfm = pd.read_csv("rfm_features.csv")

rolling = pd.read_csv("rolling_statistics.csv")
```

```python
st.set_page_config(page_title="RetailPulse Dashboard",
                   layout="wide")

st.title("📊 RetailPulse - Real-Time Metrics & Alerts")
```

```python
st.write("Last Updated")

st.success(datetime.now().strftime("%d-%m-%Y %H:%M:%S"))
```

```python
if st.button("🔄 Refresh Dashboard"):
    st.rerun()
```

```python
st.header("Forecast Metrics")

col1, col2, col3, col4 = st.columns(4)

col1.metric(
    "Total Forecast",
    int(forecast["Hybrid_Prediction"].sum())
)

col2.metric(
    "Average Forecast",
    round(forecast["Hybrid_Prediction"].mean(), 2)
)

col3.metric(
    "Maximum Forecast",
    int(forecast["Hybrid_Prediction"].max())
)

col4.metric(
    "Minimum Forecast",
    int(forecast["Hybrid_Prediction"].min())
)
```

```python
if forecast["Hybrid_Prediction"].max() > 500:
    st.warning("⚠ High Demand Detected")
else:
    st.success("Demand is Normal")
```

```python
st.subheader("Hybrid Forecast")

st.line_chart(forecast.set_index("Date")["Hybrid_Prediction"])
```

```python
st.subheader("Prophet Forecast")

st.line_chart(prophet["yhat"])
```

```python
if prophet["yhat"].iloc[-1] > prophet["yhat"].mean():

    st.info("Demand Expected to Increase")

else:

    st.success("Demand Stable")
```

```python
st.header("Inventory Metrics")

col1, col2, col3 = st.columns(3)

col1.metric(
    "Average Inventory",
    round(inventory["Recommended_Inventory"].mean(), 2)
)

col2.metric(
    "Maximum Inventory",
    int(inventory["Recommended_Inventory"].max())
)

col3.metric(
    "Minimum Inventory",
    int(inventory["Recommended_Inventory"].min())
)
```

```python
st.subheader("Inventory Alerts")

low_stock = inventory[
    inventory["Inventory_Status"] == "Low Stock"
]

if len(low_stock) > 0:
    st.error(f"⚠ {len(low_stock)} products are Low Stock.")
else:
    st.success("✅ Inventory is Healthy.")
```

```python
st.header("Customer Churn")

col1, col2, col3 = st.columns(3)

col1.metric(
    "Total Customers",
    len(churn)
)

col2.metric(
    "Average Churn Probability",
    f"{churn['ChurnProbability'].mean()*100:.2f}%"
)

col3.metric(
    "High Risk Customers",
    len(churn[churn["ChurnProbability"] > 0.80])
)
```

```python
st.subheader("Customer Alerts")

high_risk = churn[churn["ChurnProbability"] > 0.80]

if len(high_risk) > 0:
    st.error(f"⚠ {len(high_risk)} High Risk Customers Found")
else:
    st.success("✅ No High Risk Customers")
```

```python
st.header("Customer Segmentation")

col1,col2,col3 = st.columns(3)

col1.metric("Customers",
            len(rfm))

col2.metric("Average Frequency",
            round(rfm["Frequency"].mean(),2))

col3.metric("Average Monetary",
            round(rfm["Monetary"].mean(),2))
```

```python
st.header("Rolling Statistics")

st.line_chart(
rolling[["RollingMean7",
         "RollingStd7"]]
)
```

```python
if rolling["RollingStd7"].max()>100:

    st.warning("Sales Volatility Increasing")

else:

    st.success("Sales Stable")
```
